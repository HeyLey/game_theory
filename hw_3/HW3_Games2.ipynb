```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  

from IPython.core.pylabtools import figsize
figsize(7, 7)

from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
```

```python
m = np.array(
    [[-1, 3,  6, 0,  4, 1],
     [ 5, 1, -2, 4, -1, 0]])
```

```python
figsize(10,10)

fig, ax = plt.subplots()
patches = []

ax.set_xlim([0, 1.5])
ax.set_ylim([-4, 8.])

colors = ["k-", "g-", "r-", "c-", "m-", "y-"]

for i in range(6):
    plt.plot([0, 1], [m[1, i], m[0, i]], colors[i])

plt.show()
```

```python
figsize(10,10)

fig, ax = plt.subplots()
patches = []

ax.set_xlim([0, 1.5])
ax.set_ylim([-4, 8.])

for i in [0, 5]:
    plt.plot([0, 1], [m[1, i], m[0, i]], colors[i])

plt.show()
```

```python
A_ub = np.concatenate((-m.T, np.ones((6, 1))), axis=1)
b_ub = np.zeros(6)


A_eq = np.ones((1, 3))
A_eq[0, 2] = 0
b_eq = np.ones(1)

c = np.zeros(3)
c[2] = -1
```

```python
from scipy.optimize import linprog

linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq)
```

```python
A_ub = np.concatenate((m, -np.ones((2, 1))), axis=1)
b_ub = np.zeros(2)


A_eq = np.ones((1, 7))
A_eq[0, 6] = 0
b_eq = np.ones(1)

c = np.zeros(7)
c[6] = 1
```

```python
from scipy.optimize import linprog

linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq)
```
