```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
import scipy.io
import scipy.sparse
from scipy.optimize import linprog
```

```python
data = scipy.io.mmread("./data/TinySampleTrainingSet.mtx")
test_data = scipy.io.mmread("./data/TinySampleTestSet.mtx")
```

```python
m, n = data.shape

k = 5

P = np.random.rand(m, k) + 1
Q = np.random.rand(k, n) + 1
```

```python
def mse(m):
    diff = 0.0

    I, J, V = scipy.sparse.find(m)
        
    for i, j, v in zip(I, J, V):
        p = np.dot(P[i, :], Q[:, j])
        diff += (p - v) ** 2
    return diff / len(V)
```

```python
def make_step(lmb):
    for i0 in range(m):    
        diff = 0.0
        A = lmb * np.eye(k)
        b = np.zeros(k)
    
        for i, j, v in zip(*scipy.sparse.find(data)):
            if i != i0:
                continue
            A += np.outer(Q[:, j], Q[:, j])
            b += v * Q[:, j] 
        
        res = np.linalg.solve(A, b)
        P[i0, :] = res
    
    for j0 in range(n):    
        diff = 0.0
        A = lmb * np.eye(k)
        b = np.zeros(k)
    
        for i, j, v in zip(*scipy.sparse.find(data)):
            if j != j0:
                continue
            A += np.outer(P[i, :], P[i, :])
            b += v * P[i, :] 
        
        res = np.linalg.solve(A, b)
        Q[:, j0] = res
```

```python
for lmb in [0.001, 0.01, 0.1, 1]:
    P = np.random.rand(m, k) + 1
    Q = np.random.rand(k, n) + 1

    train_errors = np.zeros(50)
    test_errors = np.zeros(50)

    
    prev_val = mse(data)
    for i in range(50):
        make_step(lmb)
        val = mse(data)
        prev_val = val
        
        train_errors[i] = mse(data)
        test_errors[i] = mse(test_data)
    
    print(lmb)
    print("Train: ", mse(data)) 
    print("Test: ", mse(test_data))
    
    train_plt, = plt.plot(train_errors, label='train')
    test_plt, = plt.plot(test_errors, label='test')
    plt.legend(handles=[train_plt, test_plt])
    plt.show()
```

```python
from scipy.sparse.linalg import svds

S = np.zeros((50, 20))
```

```python
def mse(m, S):
    I, J, V = scipy.sparse.find(m)

    return np.sum(np.power((S[I, J] - V), 2)) / len(V)
```

```python
from scipy import sparse

def fit_with_gamma(gamma):
    S = np.zeros((50, 20))

    train_errors = np.zeros(50)
    test_errors = np.zeros(50)

    for k in range(50):
        I, J, V = scipy.sparse.find(data)
        diff = (S[I, J] - V)
        diff_m = sparse.coo_matrix((diff, (I,J)), shape=(50, 20))
        U, s, V = svds(diff_m, 1)

        S1 = -gamma * (U * V)
    
        p = 2 / (2 + k)

        S = S + p * (S1 - S)
        
        train_errors[k] = mse(data, S)
        test_errors[k] = mse(test_data, S)
    
    print("Train: ", mse(data, S)) 
    print("Test: ", mse(test_data, S))
    train_plt, = plt.plot(train_errors, label='train')
    test_plt, = plt.plot(test_errors, label='test')
    plt.legend(handles=[train_plt, test_plt])
```

```python
fit_with_gamma(1)
```

```python
fit_with_gamma(10)
```
